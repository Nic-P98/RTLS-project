```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('data_in_area.csv')
```

```python
data
```

```python
xv = [51.46095062608643, 51.46093088216246, 51.46025271647916,51.46028671350568,51.46034879496703,51.46038279192198,
      51.46055573256162, 51.46051138887036, 51.46095062608643]
yv = [-0.9325233041095033, -0.9322321591246415, -0.9323544769902165, -0.9328170921018635, -0.9328123473314877, -0.9332986862950139,
 -0.9332654729023827, -0.9326059498201374, -0.9325233041095033]
plt.scatter(data['lng'],data['lat'])
plt.plot(yv,xv)
```

```python
L = min(yv)-max(yv)
W = max(xv)-min(xv)

Lgrid = np.linspace(L,0,100)
Wgrid = np.linspace(0,W,100)
Linterval = L/100
Winterval = W/100
```

```python
point = zip(data['lat'],data['lng'])
result = []
z_w = min(xv)
z_l = min(yv)
lat_l,lng_l = [],[]
for i in point:
    lat_n = int((i[0]-z_w)/Winterval)
    lng_n = int((z_l-i[1])/Linterval)
    lat_l.append(lat_n)
    lng_l.append(lng_n)
    txt = "G{0}-{1}".format(lat_n,lng_n)
    result.append(txt)

data['Grid'] = result
```

```python
plt.scatter(lng_l,lat_l)
```
